```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

label_list = {"0":"neutral", "1":"happiness", "2":"surprise","3":"sadness", "4":"anger", "5":"disgust", "6":"fear","7":"contempt"}
df = pd.read_csv("../model/mytrain/2021-10-20-11:43:56/ferplus_mn_score.txt", sep=' ', header=None)
pic_dir = "./FERPlus_reshape_crop_folder/Test"
def visualize(duibi):
    font={	'color': 'red',
		'size': 15,
		'family': 'Times New Roman',}
    for key in label_list.keys():
        i = 0
        print(duibi[0] + " : ",label_list[key])
        for index,row in df.iterrows():
            if duibi[0] == "orgin":
                orgin_label = row[0][0]
                output_label = row[1]
            else:
                orgin_label = row[1]
                output_label = row[0][0]
            if int(orgin_label) != int(output_label) and int(orgin_label)==int(key):
                imgs = mpimg.imread(pic_dir+os.sep+row[0][:-4]+os.sep+"0.jpg")
    #             print("orgin_label: ",label_list[str(orgin_label)])
    #             print("output_label: ",label_list[str(output_label)])
                i+=1
                plt.subplot(3,3,i)
                plt.imshow(imgs),plt.axis("off")
                plt.text(0.1, 0.1,label_list[str(output_label)], fontdict=font)
            if i > 8:
                break
        plt.show()
```

```python
duibi=["orgin","output"]
visualize(duibi)
```
